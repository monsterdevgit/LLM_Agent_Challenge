```python
import json
import os
from helpers.data_loader import load_datasets
from helpers.user_utils import get_user_data
from helpers.config import get_client
from models.user_profiles import build_synthetic_user
from prediction.llm_simulator import simulate_user_response
```

```python
if __name__ == "__main__":

    # Load datasets
    users_df, reviews_df = load_datasets()

    # Gemini client
    client = get_client()

    results = []

    # Process users
    for _, row in users_df.head(5).iterrows():

        user_id = row["usuario"]

        # Build synthetic user (USES ALL PROFILE FUNCTIONS)
        user_features, history = build_synthetic_user(row, reviews_df)

        # Unseen product
        product = "Wireless Bluetooth headphones with noise cancellation"

        # LLM simulation
        prediction = simulate_user_response(
            client,
            user_features,
            history,
            product
        )

        results.append({
            "user": user_id,
            "product": product,
            "prediction": prediction,
            "user_features": user_features, 
            "history": history
        })

        print(f"[INFO] Processed user: {user_id}")

    # Ensure directory exists
    os.makedirs("data/processed", exist_ok=True)

    # Save results (ONLY ONCE )
    with open("data/processed/predictions.json", "w", encoding="utf-8") as f:
        json.dump(results, f, indent=2, ensure_ascii=False)

    print("\n Simulation complete and saved.")
```
